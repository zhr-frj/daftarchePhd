```python
# --- ุณุงุฎุช ยซุฎุท ููููยป ู ุงุจุฒุงุฑูุง LangChain ---
print("--- ุฏุฑ ุญุงู ุณุงุฎุช ยซุฎุท ููููยป (Pipeline) ---")
text_gen_pipeline = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=512  # ุงูุฒุงุด ุทูู ูพุงุณุฎ ุจุฑุง ุฌูุงุจโูุง ฺฉุงููโุชุฑ
)

# ุณุงุฎุช ยซุบูุงูยป LangChain ุจุฑุง LLM
llm = HuggingFacePipeline(pipeline=text_gen_pipeline)

# ุณุงุฎุช ยซุชูุฒฺฉููุฏู ุฎุฑูุฌยป
parser = StrOutputParser()

print("--- ยซุฎุท ููููยปุ ยซุบูุงูยป ู ยซุชูุฒฺฉููุฏูยป ุขูุงุฏู ุดุฏูุฏ ---")
```

```python
# --- ุชุงุจุน ฺฉูฺฉ ุจุฑุง ุงุนูุงู ยซูุงูุจ ฺุชยป ---

# ุงู ุชุงุจุนุ ูพุฑุงููพุช ููุง ูุง ุฑุง ุฏุฑ ูุงูุจ ฺุช ุฑุณู ูุฏู Gemma ูุฑุงุฑ ูโุฏูุฏ
def apply_chat_template_and_response(prompt):
    messages = [
    {'role': 'user', 'content': prompt}
    ]

    # ุงุนูุงู ูุงูุจ ฺุช
    text = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True,
        enable_thinking=False
    )

    # ูุฑุงุฎูุงู ูุฏู ู ุชูุฒฺฉุงุฑ ุงููู
    response = llm.invoke(text)
    # response ุฑุง ุชูุฒ ูโฺฉูู ุชุง ููุท ุฌูุงุจ ููุง ุจุงู ุจูุงูุฏ
    return response.replace(text, '').strip()

print("--- ุชุงุจุน ฺฉูฺฉ ยซูุงูุจ ฺุชยป ุขูุงุฏู ุดุฏ ---")
```
